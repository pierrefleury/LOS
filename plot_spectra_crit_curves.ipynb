```python
%matplotlib inline
import sys, platform, os
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
# Use TeX
from matplotlib import rc
# rc('font',**{'family':'serif','serif':['Times']})
rc('text', usetex=True)
rc('font', family='serif')
#rc('font', **{'serif':['Times']})
matplotlib.rcParams.update({'font.size': 15})

import numpy as np

# Changing radians to arcsec
rad2arcsec=648*1000/np.pi
```

```python
#Loading the spectra

P22=np.loadtxt('./data_spectra/P22.txt', dtype=float)
P23=np.loadtxt('./data_spectra/P23.txt', dtype=float)
```

```python
#PDFs for the various string lens parameters

def probln(m,s,x):
    "log-normal proba density distrib"
    proba=np.exp(-((np.log(x)-m)**2)/(2*s**2))/(np.sqrt(2*np.pi)*s*x)
    return proba

def probgauss(m,s,x):
    "Gaussian proba density distrib"
    proba=np.exp(-(x-m)**2/(2*s**2))/(np.sqrt(2*np.pi)*s)
    return proba

def probray(s,x):
    "Rayleigh distribution"
    proba=x*np.exp(-x**2/(2*s**2))/s**2
    return proba
```

```python
# Plot the pdfs for the parameter used.
sEr=0.5 # 
mEr=np.log(0.8)-sEr**2+np.log(np.pi*0.001/648)   #using radians
szd=0.5
muzs=2
sigzs=0.6
tt=np.linspace(0.001,3,100)/rad2arcsec  # thetaE in arsec the converted in radians
zz=np.linspace(0,5,100)
Ptheta=probln(mEr,sEr,tt)
Pzs=probgauss(muzs,sigzs,zz)
Pzd=probray(szd,zz)

fig, sub=plt.subplots(1,2,sharey=False,figsize=(8,4))

sub[0].plot(tt*rad2arcsec,Ptheta/rad2arcsec,color='black')
sub[0].set_xlabel(r'$\theta_{\rm E}$ [arcsec]')
sub[0].set_ylabel(r'$p\left(\theta_{\rm E}\right)$')
sub[0].set_xticks((0,0.5,1,1.5,2,2.5,3))
sub[0].axis([0, 3, 0, 1.5])
#sub[0].set_aspect('equal')

sub[1].plot(zz,Pzs,color='blue')
sub[1].plot(zz,Pzd,color='red')
sub[1].set_xlabel(r'$z$')
sub[1].set_ylabel(r'$p\left(z\right)$')
sub[1].axis([0, 5, 0, 1.3])
plt.legend([r'$p_{s}\left(z_{s}\right)$',r'$p_{d}\left(z_{d}\right)$'], frameon=False);

fig.tight_layout(pad=2)
#fig.savefig('./figures/pdfs_used_new.png', format='png', transparent=True, bbox_inches='tight', pad_inches=0.1)
```

```python
# Now plotting the spectra


# Power laws and kappaE used
pl=np.linspace(1.5,2.5,5)
kappaE=3/2-pl/2

ll=np.exp(np.log(10)*np.linspace(0,4,150))

fig, ax=plt.subplots(nrows=2,
                     ncols=2,
                     sharey=False,
                     sharex=True,
                     figsize=(10,6),
                     gridspec_kw=dict(height_ratios=[2,1]))
fig.subplots_adjust(hspace=0)


# Spectra for n1=2, n2=2

# Absolute spectrum
ax[0,0].loglog(ll,P22[:,0])
ax[0,0].loglog(ll,P22[:,2],color='black')
ax[0,0].loglog(ll,P22[:,4])
ax[0,0].axvspan(0, 2, alpha=0.3, color='grey')
ax[0,0].axvspan(450, 10000, alpha=0.3, color='grey')
ax[0,0].set_ylabel(r"$P_{n_1 n_2}(\ell)$")
ax[0,0].set_ylim([2e-13,3e-9])
ax[0,0].set_title(r"$(n_1, n_2)=(2,2)$")

# Relatively to the SIS
ax[1,0].semilogx(ll,P22[:,0]/P22[:,2])
ax[1,0].semilogx(ll,P22[:,4]/P22[:,2])
ax[1,0].semilogx(ll,P22[:,2]/P22[:,2],color='black')
ax[1,0].axvspan(0, 2, alpha=0.3, color='grey')
ax[1,0].axvspan(450, 10000, alpha=0.3, color='grey')
ax[1,0].set_ylim([0.95,1.05])
ax[1,0].set_xlim([1,1e4])
ax[1,0].set_xlabel(r"$\ell$")
ax[1,0].set_ylabel(r"$P_{n_1 n_2}(\ell)/P_{n_1 n_2}^{\rm SIS}(\ell)$")


# Spectra for n1=2, n2=3
# Absolute spectrum
ax[0,1].loglog(ll,P23[:,0],
               label=r'$\nu=3/2 \Leftrightarrow \kappa_{\rm E}=3/4$')
ax[0,1].loglog(ll,P23[:,4],
               label= r'$\nu=5/2 \Leftrightarrow \kappa_{\rm E}=1/4$')
ax[0,1].loglog(ll,P23[:,2], color='black',
               label=r'$\nu=2\,(\mathrm{SIS}) \Leftrightarrow \kappa_{\rm E}=1/2$')
ax[0,1].axvspan(0, 2, alpha=0.3, color='grey')
ax[0,1].axvspan(450, 10000, alpha=0.3, color='grey')
ax[0,1].set_ylim([2e-16,3e-13])
ax[0,1].set_title(r"$(n_1, n_2)=(2,3)$")
ax[0,1].legend(frameon=False, loc='lower right');

# Relatively to the SIS
ax[1,1].semilogx(ll,P23[:,0]/P23[:,2])
ax[1,1].semilogx(ll,P23[:,4]/P23[:,2])
ax[1,1].semilogx(ll,P23[:,2]/P23[:,2], color='black')
ax[1,1].axvspan(0, 2, alpha=0.3, color='grey')
ax[1,1].axvspan(450, 10000, alpha=0.3, color='grey')
ax[1,1].set_ylim([0.9,2.5])
ax[1,1].set_xlim([1,1e4])
ax[1,1].set_xlabel(r"$\ell$");
                   
    
#fig.savefig('figures/spectra_critical_modes.pdf',
#            format='pdf',
#            transparent=True,
#            bbox_inches='tight',
#            pad_inches=0.1)
```
